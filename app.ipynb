```python
print("\nUnique values in 'Social Media Platforms':")
unique_values = data['Social Media Platforms'].unique()
print(unique_values)
print(f"Number of unique values: {len(unique_values)}")
```

```python
# Code to count mentions of each social media platform
# Define the list of platforms to count
platforms = ['Facebook', 'Instagram', 'YouTube', 'Twitter', 'Discord', 'Reddit', 'TikTok', 'Snapchat', 'Pinterest']

# Initialize a dictionary to store the counts for each platform
platform_counts = {platform: 0 for platform in platforms}

# Iterate over each row in the 'Social Media Platforms' column
for platforms_in_row in data['Social Media Platforms']:
    # Split the platforms string by commas and strip spaces
    platforms_list = [platform.strip() for platform in platforms_in_row.split(',')]
    
    # Count the occurrences of each platform in the row
    for platform in platforms_list:
        if platform in platform_counts:
            platform_counts[platform] += 1

# Print the counts for each platform
print("\nSocial Media Platform Counts:")
for platform, count in platform_counts.items():
    print(f"{platform}: {count}")
```

```python
# Code to create the new columns
# Define the list of platforms
platforms = ['Facebook', 'Instagram', 'YouTube', 'Twitter', 'Discord', 'Reddit', 'TikTok', 'Snapchat', 'Pinterest']

# Create a new column for each platform with "Yes" or "No"
for platform in platforms:
    # Check if the platform is in the 'Social Media Platforms' column
    data[platform] = data['Social Media Platforms'].apply(lambda x: 'Yes' if platform in x else 'No')

# Display the first few rows to verify the new columns
print("\nData with new platform columns:")
print(data[['Gender', 'Social Media Platforms'] + platforms].head())  # Displaying a few columns as an example
```

```python
# Check for outliers in numerical columns (e.g., 'Age')
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers in 'Age'
outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]
print(f"\nNumber of outliers detected in 'Age': {len(outliers)}")
print(outliers)
```

```python
print(data.head())
```

```python
# Cap the outliers in 'Age' to the lower and upper bounds
data['Age'] = data['Age'].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

# Check the updated data
print("\nUpdated data with capped outliers in 'Age':")
print(data['Age'].describe())
```

```python
# Round the values in the 'Age' column to the nearest integer and convert to int64
data['Age'] = data['Age'].round().astype('int64')

# Verify the changes
print("\nUpdated 'Age' column:")
print(data['Age'].head())
```

```python
data["Age"].value_counts()
```

```python
# Convert True/False columns (platform columns) to numeric (1 for True, 0 for False)
platform_columns = ['Facebook', 'Instagram', 'YouTube', 'Twitter', 'Discord', 'Reddit', 'TikTok', 'Snapchat', 'Pinterest']
for platform in platform_columns:
    data[platform] = data[platform].astype(int)

# Verify the changes
print("\nUpdated Data Types after conversion to 1/0:")
print(data[platform_columns].head())
```

## Statistical Tests (T-tests)

### T-test for Facebook Usage (Restlessness from Absence)

```python
from scipy import stats
```

```python
# T-test for 'Restlessness from Absence' between Facebook users and non-users
facebook_users = data[data['Facebook'] == 1]['Restlessness from Absence']
facebook_non_users = data[data['Facebook'] == 0]['Restlessness from Absence']
```

```python
# Perform the t-test
t_stat, p_value = stats.ttest_ind(facebook_users, facebook_non_users)

# Display the results
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a significant difference in Restlessness from Absence between Facebook users and non-users.")
else:
    print("There is no significant difference in Restlessness from Absence between Facebook users and non-users.")
```
